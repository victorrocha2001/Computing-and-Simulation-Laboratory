```python
import numpy as np
import scipy
from scipy.special import gamma
from scipy.stats import chi2

# Função potencial
def potencial(dirichlet, alfa):
    return np.prod(dirichlet ** (np.array(alfa) - 1))

# Hipótese
def hip(x, ajudante):
    return -potencial([x, 1 - (x + (1 - np.sqrt(x)) ** 2), (1 - np.sqrt(x)) ** 2], ajudante)

# Função verdade
def verdade(V, fronteira, beta, quantidade):
    if V < 0:
        return -1
    k = np.searchsorted(fronteira, V * beta)
    aproximacao = 1540 * k
    if k != 2500:
        aproximacao += np.searchsorted(quantidade[k], V * beta)
    resultado = aproximacao / 3850000
    return resultado

# Função de e-valor padronizado
def std_e(e_v):
    t, h = 2, 1
    df = t - h
    e_v_b = 1 - e_v
    qq_b = chi2.cdf(chi2.ppf(e_v_b, t), df)
    qq = 1 - qq_b
    return qq

# Função corpo
def corpo(vetor_x, vetor_y):
    alpha = np.array(vetor_x) + np.array(vetor_y)

    # Constante de normalização
    beta = np.prod(gamma(alpha)) / gamma(np.sum(alpha))

    # Gerando os thetas dirichlet
    np.random.seed(13686431)
    dirichlet = np.random.dirichlet(alpha, size=3850000)

    # Calculando valores potenciais
    valores = np.apply_along_axis(potencial, 1, dirichlet, alpha)
    valores.sort()

    # Criando as divisões
    quantidade = np.array_split(valores, 2500)
    fronteira = [q[-1] for q in quantidade]

    # Maximizando a hipótese
    global ajudante
    ajudante = alpha
    res = scipy.optimize.minimize_scalar(fun=lambda x: hip(x, ajudante), bounds=(0.0, 1.0), method='bounded')

    # Calculando e-valor e e-valor padronizado
    e_valor = verdade(-res.fun / beta, fronteira, beta, quantidade)
    std = std_e(e_valor)

    # Testando a hipótese
    if std < 0.05:
        print("Hipótese rejeitada")
        return 0
    else:
        print("Hipótese não rejeitada")
        return 1

# Vetores x e vetores y
vx = [
    [1, 17, 2], [1, 16, 3], [1, 15, 4], [1, 14, 5], [1, 13, 6], [1, 12, 7], [1, 11, 8],
    [1, 10, 9], [1, 9, 10], [1, 8, 11], [1, 7, 12], [1, 6, 13], [1, 5, 14], [1, 4, 15],
    [1, 3, 16], [1, 2, 17], [1, 1, 18], [5, 14, 1], [5, 13, 2], [5, 12, 3], [5, 11, 4],
    [5, 10, 5], [5, 9, 6], [5, 8, 7], [5, 7, 8], [5, 6, 9], [5, 5, 10], [9, 10, 1],
    [9, 9, 2], [9, 8, 3], [9, 7, 4], [9, 6, 5], [9, 5, 6], [9, 4, 7]
]
vy = [[0, 0, 0], [1, 1, 1]]

# Calculando a quantidade total e de cada tipo de hipóteses aceitas
cont, tipo_0, tipo_1 = 0, 0, 0
for vx_item in vx:
    for vy_item in vy:
        cont += 1
        if vy_item == vy[0]:
            tipo_0 += corpo(vx_item, vy_item)
        else:
            tipo_1 += corpo(vx_item, vy_item)

print(f"Não foram rejeitados {tipo_0 + tipo_1} do total de {cont}")
print(f"Não foram rejeitados {tipo_0} quando o vetor Y é {vy[0]}")
print(f"Não foram rejeitados {tipo_1} quando o vetor Y é {vy[1]}")
```
